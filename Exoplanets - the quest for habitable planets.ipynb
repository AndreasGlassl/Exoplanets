### Introduction

This project entails the analysis of the data acquired by space scientists to answer the question: *is there any planet habitable outside our solar system?*

The focus is on the analysis of measured or calculated parameters that can be relatable to the conditions observed for the Planet Earth and its relationship with the Sun. These parameters can be:

- **Planet orbit**
- **Mass of the star** 
- **Distance between the planet and the star**
- **Planet temperature**
- **Atmosphere**

Moreover, based on the acquired data, the distance from Planet Earth to the identified possible candidates can be calculated, as well as the estimated travel time by the current technologies. 

### Data acquisition
    
  - 1. NASA Exoplanet Archive (USA): Data for 4,144 confirmed exoplanets was obtained through API. The sample query included the name of the table (`exoplanets`) and the desired columns (`default` and `planet`). The format of the data was as a csv file. https://exoplanetarchive.ipac.caltech.edu/docs/program_interfaces.html



  - 2. NASA Exoplanet Archive (USA): Data regarding the description of the column names ("Table Label") was obtained by web scraping to add into the data frame for analysis. https://exoplanetarchive.ipac.caltech.edu/docs/API_exoplanet_columns.html



  - 3. The Extrasolar Planets Encyclopaedia (EU): Data was acquired from the European source for space data for comparison purposes. The data for 4248 confirmed planets was obtained as a csv file from the website: http://exoplanet.eu/catalog/
  

### Data analysis

To answer the question: *is there any planet habitable?*, we looked into the acquired data to retrieve information regarding the parameters to match possible planets that may resemble the Planet Earth. The found parameters were **planet orbit**, **mass of the star** and **planet temperature**. In each case, data was found for only a fraction of the total confirmed planets. Therefore, our analysis was narrowed down to those. 

Initially, we analyzed the the relationship between the planet orbit and the mass of the star. The data was narrowed down to the planets with an orbit lying within a range of 20% of that of Earth's orbit. Similarly, the mass of the respective star was narrowed down to a 20% range of the Sun's mass. A total of 22 planets was found.

Subsequently, the temperature of the planet was analyzed, and the data was narrowed down to the range of Earth's temperature, with a lower limit at -60 °C (as observed in Antartica) and an upper limit at 60 °C (recorded once in California). 28 planets were found. The data was is showed in terms of the distance from the Earth.

Finally, from the pool of options, two planets were found by filtering the previous set of 22 planets to those having an orbit close to that of the Earth. From the two, only Kepler-452b is estimated as habitable due to the size of the planet, which resembles that of the Planet Earth. The analysis was corroborated with NASA. Indeed, the NASA describes Kepler-452b as the most Earth-like planet.



This analysis serves as a first approximation to quest for habitable planets beyond our solar system. However, more parameters should be considered in the study. The existence of an atmosphere and its composition is essential to further prove the suitability of any planet to be habitable. But despite the progress in the field since the discovery of the first exoplanet four decades ago, there is no available technology that can provide information about the presence of an atmosphere for planets at a distance beyond our solar system.
